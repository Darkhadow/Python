```python
import pandas as pd

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)

cus_df = pd.read_csv("customer-related_20201125.csv")
prds_df = pd.read_csv("product_related_dataset_20201125.csv")
```

```python
cus_df.head()
```

```python
prds_df.head()
```

```python
cus_df.drop(['username','Unnamed: 0','gender'],axis = 1,inplace = True)
cus_df.head()
```

```python
cus_df['age'] = cus_df.age.astype(float)
cus_df.head()
```

_________
## How to round off numbers using `.round(decimals = [num])`

```python
prds_df['Pricing'] = prds_df.Pricing.round(decimals = 2)
prds_df.head()
```

_____
## Below is a test line to see if rows have been removed or not
### It should be empty

```python
prds_df[prds_df['Pricing'] < 2.0]
```

```python
prds_df = prds_df.loc[:,['main_category','product_name','Pricing','quantity']]
prds_df.head()
```

____
## How to:
### 1. Filter out rows base on given conditions
### 2. Remove said rows from df

```python
below_2 = prds_df[(prds_df['Pricing']<2.0)  |  (prds_df['main_category']=='Miscellaneous')].index
prds_df.drop(below_2, inplace = True)
prds_df.head()
```
